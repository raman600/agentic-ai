```python
import asyncio
from autogen_agentchat.ui import Console
from autogen_agentchat.agents import AssistantAgent, SocietyOfMindAgent, UserProxyAgent
from autogen_ext.models.openai import OpenAIChatCompletionClient
from autogen_agentchat.teams import RoundRobinGroupChat
from autogen_agentchat.conditions import TextMentionTermination

async def main() -> None:
    model_client = OpenAIChatCompletionClient(model="gpt-4o",api_key="api_key")

    agent1 = AssistantAgent(
        "planner", 
        model_client=model_client, 
        system_message="You are a software project planner. \
            Break down coding requests into clear, actionable steps. Focus on planning and requirements clarification."
    )
    agent2 = AssistantAgent(
        "coder",
        model_client=model_client,
        system_message="You are a software engineer. Implement code based on the planner's specifications.\
             Write clean, well-documented code with proper error handling.",
    )
    agent3 = AssistantAgent(
        "reviewer",
        model_client=model_client,
        system_message="You are a code reviewer.\
             Evaluate code quality, correctness, and best practices. Provide specific feedback for improvements.\
             When the solution is complete and meets all requirements, respond with 'APPROVE'.",
    )
    
    termination = TextMentionTermination("APPROVE")

    inner_team = RoundRobinGroupChat([agent1, agent2, agent3], termination_condition=termination)

    society_of_mind_agent = SocietyOfMindAgent(
        "society_of_mind", 
        team=inner_team, 
        model_client=model_client,
        response_prompt='Provide a complete, well-structured solution to the user request.\
             Include all code, explanations, and any additional context needed.\
             Present this as a unified response without mentioning the internal collaboration process.'
    )

    user_proxy_agent = UserProxyAgent(
        name ='HumanProxy',
        description='A human user who can provide feedback, ask questions, and approve or request modifications to the solution.',
        input_func=input
    )

    outer_team = RoundRobinGroupChat([society_of_mind_agent, user_proxy_agent],termination_condition=termination)

    stream = outer_team.run_stream(task="Write a Python function that takes two numbers and returns their sum. Include one test case.")
    await Console(stream)

await(main())


```
